```python
import pyautogui
import time
import pyperclip
import os

pyautogui.FAILSAFE = True


NEW_CHAT = (107, 249)
TEXTBOX_FIRST = (802, 547)
TEXTBOX_SECOND = (787, 953)
PARAGRAPH = (815, 389)


def send(surah_number, ayah_start, ayah_end, sys_prompt, user_prompt):
    PROMPT_1 = sys_prompt
    PROMPT_2 = user_prompt


    LOAD_WAIT = 2
    RESPONSE_WAIT_1 = 15
    RESPONSE_WAIT_2 = 30


    folder_name = "tafaseer"
    os.makedirs(folder_name, exist_ok=True)

    filename = f"{surah_number}_{ayah_start}-{ayah_end}.txt"
    filepath = os.path.join(folder_name, filename)


    print("Starting in 5 seconds...")
    time.sleep(5)

    # New chat
    pyautogui.click(NEW_CHAT)
    time.sleep(LOAD_WAIT)

    # First prompt
    pyautogui.click(TEXTBOX_FIRST)
    pyperclip.copy(PROMPT_1)
    pyautogui.hotkey("ctrl", "v")
    pyautogui.press("enter")

    time.sleep(RESPONSE_WAIT_1)

    # Second prompt
    pyautogui.click(TEXTBOX_SECOND)
    pyperclip.copy(PROMPT_2)
    pyautogui.hotkey("ctrl", "v")
    pyautogui.press("enter")

    time.sleep(RESPONSE_WAIT_2)

    # Select paragraph (triple click)
    pyautogui.click(PARAGRAPH, clicks=3, interval=0.2)
    time.sleep(0.5)

    # Copy
    pyautogui.hotkey("ctrl", "c")
    time.sleep(0.5)

    # Get copied text
    copied_text = pyperclip.paste()
    print(copied_text)
    # Save to file
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(copied_text)

    print(f"Saved output to {filepath}")
```
