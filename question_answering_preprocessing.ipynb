```python
from datasets import load_dataset
import pandas as pd
import numpy as np

```

```python
ds = load_dataset("nazimali/quran-question-answer-context")
```

```python
ds
```

```python
df = ds["train"].to_pandas()
df
```

```python
df.columns.tolist()
```

```python
"""
columns needed:

1- question_en
2- answer_en
3- chapter_no
4- verse_list
"""
```

```python
qa_df = df[["question_en", "answer_en", "chapter_no", "verse_list"]].copy()
```

```python
qa_df
```

```python
qa_df.isnull().sum()
```

```python
text = qa_df.loc[30, "question_en"]
print(repr(text))
```

```python
qa_df["question_en"] = (
    qa_df["question_en"]
    .replace(r"^\s*$", np.nan, regex=True)
)
```

```python
qa_df.isnull().sum()
```

```python
qa_df = qa_df.dropna(subset=["question_en"])
```

```python
qa_df
```

```python
i = 25
print(qa_df['question_en'][i])
print("_____________________")
print(qa_df['answer_en'][i])
print("_____________________")
print(qa_df['chapter_no'][i], qa_df['verse_list'][i])
```

```python
qa_df.loc[:, "answer_en"] = qa_df["answer_en"].str.replace(r"\(.*?\)", "", regex=True).str.strip()
qa_df.loc[:, "question_en"] = qa_df["question_en"].str.replace(r"\(.*?\)", "", regex=True).str.strip()
```

```python
qa_df.to_csv("question_answering.csv", encoding="utf-8")
```
